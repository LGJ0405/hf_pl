```python
from transformers import pipeline
import random
import re

model_name = "Qwen/Qwen2.5-7B"
generator = pipeline("text2text-generation", model=model_name)
```

```python
# 알고리즘
KNOWN_ALGOS = [
    "깊이 우선 탐색", "너비 우선 탐색", "DFS", "BFS",
    "동적 프로그래밍", "DP",
    "그리디", "탐욕법",
    "이진 탐색", "Binary Search",
    "최단 경로", "다익스트라", "벨만-포드", "플로이드-워셜",
    "정렬", "분할 정복", "트리", "그래프", "백트래킹"
]

def recommend_algorithms():
    # 무작위로 5개 추천
    recommendations = random.sample(KNOWN_ALGOS, 5)
    print("추천 알고리즘 유형 5개 (무작위 순서):")
    for i, algo in enumerate(recommendations, 1):
        print(f"{i}. {algo}")
    return recommendations

def is_known_algorithm(user_input):
    # 존재 여부 확인
    return any(algo.lower() in user_input.lower() for algo in KNOWN_ALGOS)
```

```python
# 알고리즘 - 문제 유형 매핑
ALGO_TO_PROBLEM_TYPE = {
    "DFS": "그래프 탐색",
    "BFS": "그래프 탐색",
    "깊이 우선 탐색": "그래프 탐색",
    "너비 우선 탐색": "그래프 탐색",
    "트리": "그래프 탐색",
    "그래프": "그래프 탐색",

    "동적 프로그래밍": "최적화/점화식",
    "DP": "최적화/점화식",
    "백트래킹": "최적화/점화식",

    "그리디": "탐욕법 문제",
    "탐욕법": "탐욕법 문제",

    "이진 탐색": "정렬 배열 탐색",
    "Binary Search": "정렬 배열 탐색",
    "정렬": "정렬 배열 문제",
    "분할 정복": "분할 정복 문제",

    "최단 경로": "그래프 최단 경로",
    "다익스트라": "그래프 최단 경로",
    "벨만-포드": "그래프 최단 경로",
    "플로이드-워셜": "그래프 최단 경로"
}
```

```python
# 추천 알고리즘 출력
recommendations = recommend_algorithms()

# 테스트용 하드코딩 (실제 실행 시 input() 사용 가능)
# algo_input = input('번호 입력: ')
algo_input = '1'

# 입력값 처리
if algo_input.isdigit():
    idx = int(algo_input)
    algo = recommendations[idx - 1] if 1 <= idx <= len(recommendations) else recommendations[0]
else:
    algo = algo_input if is_known_algorithm(algo_input) else recommendations[0]
```

```python
print(f"{algo} 알고리즘 문제 생성 중 **\n")

problem_type = ALGO_TO_PROBLEM_TYPE.get(algo)

# 프롬프트 파일 읽기
with open("./prompt_template.txt", "r", encoding="utf-8") as f:
    template = f.read()
prompt = template.replace("{algo_name}", algo)
prompt = prompt.replace("{problem_type}", problem_type)

# 문제 생성
result = generator(prompt, max_new_tokens=600, do_sample=True, top_p=0.95, temperature=0.7)
result = result[0]['generated_text']
print('생성 완료')
print(result)
```

```python
# 프롬프트 제거 후 공백 정리
problem = result.replace(prompt, "").strip()

# 결과 출력
print("생성된 문제:")
print(problem)
```
